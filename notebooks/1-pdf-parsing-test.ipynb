```python
import pdfplumber
import os
import re
```

Extract Raw Text from the PDF

```python
pdf_folder_path = os.path.join('..', 'data', 'raw_pdfs', '2024')
pdf_filename = "JEE Main 2024 (01 Feb Shift 1) Previous Year Paper with Answer Keys - MathonGo.pdf"
file_path = os.path.join(pdf_folder_path, pdf_filename)
try:
    # This is the "mission"
    with pdfplumber.open(file_path) as pdf:
        first_page = pdf.pages[0] 
        text = first_page.extract_text()
        
        print("\n--- SUCCESSFULLY EXTRACTED TEXT FROM PAGE 1 ---")
        print(text)
except FileNotFoundError:
    print(f"\n--- ERROR ---")
    print(f"File not found at: {file_path}")
except Exception as e:
    print(f"\nAn error occurred: {e}")
```

Parse the Answer Key (Regex)

```python
pdf_folder_path = os.path.join('..', 'data', 'raw_pdfs', '2024')
pdf_filename = "JEE Main 2024 (01 Feb Shift 1) Previous Year Paper with Answer Keys - MathonGo.pdf"
file_path = os.path.join(pdf_folder_path, pdf_filename)
print(f"Attempting to open file at: {file_path}")

try:
    with pdfplumber.open(file_path) as pdf:
        # Get the LAST page (index -1)
        last_page = pdf.pages[-1] 
        text = last_page.extract_text()
        
        print("\n--- RAW TEXT FROM LAST PAGE ---")
        print(text)
        pattern = re.compile(r"(\d+)\.\s*\(([^\)]+)\)")

        matches = pattern.findall(text)

        answer_key = {question_num: answer for question_num, answer in matches}
        
        print("\n--- PARSED ANSWER KEY (DICTIONARY) ---")
        print(answer_key)
        
        print(f"\nSuccessfully parsed {len(answer_key)} answers.")

except Exception as e:
    print(f"\nAn error occurred: {e}")
```
